```python
from sklearn.datasets import load_breast_cancer
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.neighbors import KNeighborsClassifier
import seaborn as sns

import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow import keras

from sklearn.preprocessing import StandardScaler, LabelEncoder

from keras.models import Sequential
from keras.layers import Dense, Input
from keras.optimizers import SGD,Adam

df=pd.read_csv("C:\\Users\\user\\2024_ai_test\\data\\PAMAP_features.csv", index_col=0)
df
```

```python
label=LabelEncoder()
df['activity_id']=label.fit_transform(df['activity_id'])
```

```python
y=df['activity_id']#주의. 원핫인코딩 필요
y=pd.get_dummies(y).values
print(y[0])
```

```python
X=df.drop(['act_level', 'activity_id', 'sub_id'], axis=1)#주의
X.head()
```

```python
X= np.expand_dims(X, axis=2)
print(X.shape)
```

```python
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.2)#주의
print(X_train.shape)
print(y_train.shape)
```

```python
from tensorflow.keras.layers import Input, Dense, Conv1D, MaxPooling1D, Flatten, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras import optimizers
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical

# 모델 정의
inputs = Input(shape=(64,1))
hidden1 = Conv1D(32, kernel_size=3, activation='relu')(inputs)
hidden2 = MaxPooling1D(pool_size=2)(hidden1)
hidden3 = Conv1D(64, kernel_size=3, activation='relu')(hidden2)
hidden4 = MaxPooling1D(pool_size=2)(hidden3)
hidden5 = Conv1D(128, kernel_size=3, activation='relu')(hidden4)
hidden6 = MaxPooling1D(pool_size=2)(hidden5)
hidden7 = Flatten()(hidden6)
hidden8 = Dropout(0.5)(hidden7)
hidden9 = Dense(128, activation="relu")(hidden8)
output = Dense(12, activation="softmax")(hidden9)

model = Model(inputs=inputs, outputs=output)
```

```python
# 모델 컴파일
model.compile(optimizer=optimizers.Adam(learning_rate=0.001), loss='categorical_crossentropy', metrics=['accuracy'])
# 모델 요약
model.summary()
# 모델 학습
history = model.fit(X_train, y_train, epochs=5, batch_size=128, validation_split=0.2)
```

```python
test_loss, test_acc=model.evaluate(X_test, y_test)
print("테스트 정확도: ", test_acc)
```

```python
loss=history.history['loss']
acc=history.history['accuracy']
epochs=range(1, len(loss)+1)

plt.plot(epochs, loss, 'b', label="Training Loss")
plt.plot(epochs, acc, 'r', label="Accuracy")
plt.xlabel('epochs')
plt.ylabel('loss/acc')
plt.show()
```
